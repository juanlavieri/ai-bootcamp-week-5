```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = Path('./Resources/movie_scores.csv')
movies_df = pd.read_csv(csv_path)
movies_df.head()
```

```python
# Figure out the minimum and maximum IMDB user vote count
print(movies_df["IMDB_user_vote_count"].min())
print(movies_df["IMDB_user_vote_count"].max())
```

```python
# Create bins in which to place values based upon IMDB vote count
bins = [0, 2500, 5000, 10000, 20000, 30000, 50000, 100000, 350000]
bins = [int(b) for b in bins] # convert the list of strings to a list of integers
# Create labels for these bins
group_labels = ['0 to 2.4k', '2.5k to 4.9k', '5k to 9k', '10k to 14k', '15k to 19k', '20k to 29k', '30k to 49k', '50k to 99k']

```

```python
# Slice the data and place it into bins
movies_df["IMDB_user_vote_count"] = pd.cut(movies_df["IMDB_user_vote_count"], bins, labels=group_labels)
```
