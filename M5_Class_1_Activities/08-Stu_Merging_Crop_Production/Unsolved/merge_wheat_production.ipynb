```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Import data
wheat_production_2018_filepath = Path("Resources/G20_wheat_production_2018.csv")
wheat_production_2019_filepath = Path("Resources/G20_wheat_production_2019.csv")
wheat_production_2020_filepath = Path("Resources/G20_wheat_production_2020.csv")


# Read in data 
wheat_production_2018 = pd.read_csv(wheat_production_2018_filepath)
wheat_production_2019 = pd.read_csv(wheat_production_2019_filepath)
wheat_production_2020 = pd.read_csv(wheat_production_2020_filepath)
```

```python
# Display the 2018 wheat data
wheat_production_2018.head()
```

```python
# Display the 2019 wheat data
wheat_production_2019.head()
```

```python
# Display the 2020 wheat data
wheat_production_2020.head()
```

### Merge the 2018 wheat data with the 2019 wheat data. 

```python
#  Perform an inner merge that combines the 2018 and 2019 DataFrames on the `Country` column.
wheat_production_2018_2019 = pd.merge(wheat_production_2018, wheat_production_2019, on="Country", how="inner")
wheat_production_2018_2019.head()
```

```python
# Perform another inner merge that combines the 2018-2019 merged DataFrame 
# with the 2020 DataFrame on the `Country` column. 
wheat_production_2018_2019_2020 = pd.merge(wheat_production_2018_2019, wheat_production_2020, on="Country", how="inner")
wheat_production_2018_2019_2020.head()
```

```python
# Get the column names.
wheat_production_2018_2019_2020.columns
```

```python
# Drop the "Crop_x", "Year_x", "Crop_y", "Year_y", "Crop" and "Year" columns. 
wheat_production_2018_2019_2020.drop(columns=["Crop_x", "Year_x", "Crop_y", "Year_y", "Crop", "Year"], inplace=True)
wheat_production_2018_2019_2020.head()
```

```python
# Rename the remaining columns (Tonnes of Wheat HA)_2018, (Tonnes of Wheat HA)_2019, and (Tonnes of Wheat HA)_2020. 
wheat_production_2018_2019_2020.rename(columns={"Value(tonnes of HA)_x": "(Tonnes of Wheat HA)_2018", "Value(tonnes of HA)_y": "(Tonnes of Wheat HA)_2019", "Value(tonnes of HA)": "(Tonnes of Wheat HA)_2020"}, inplace=True)


# Display the DataFrame
wheat_production_2018_2019_2020.head()
```

```python
# Get the columns
wheat_production_2018_2019_2020.columns
```

```python
# Sort on the amount of tonnes of wheat for each year and reset the index. 
wheat_production_2018_2019_2020.sort_values(by=["(Tonnes of Wheat HA)_2018", "(Tonnes of Wheat HA)_2019", "(Tonnes of Wheat HA)_2020"], ascending=False, inplace=True)


# Display the sorted DataFrame.
wheat_production_2018_2019_2020.head()
```
