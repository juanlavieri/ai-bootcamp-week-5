### Import Libraries and Dependencies

```python
import pandas as pd
from pathlib import Path
```

### Read in files

```python
# Import data
customer_df = pd.read_csv(Path("Resources/customer_info.csv"), index_col="CustomerID")
france_df = pd.read_csv(Path("Resources/france_products.csv"), index_col="CustomerID")
netherlands_df = pd.read_csv(Path("Resources/netherlands_products.csv"), index_col="CustomerID")
uk_df = pd.read_csv(Path("Resources/uk_products.csv"), index_col="CustomerID")
products_df = pd.read_csv(Path("Resources/products.csv"), index_col="CustomerID")
# Read in data and make the index the CustomerID
customer_df.head()


```

### Concatenate data by rows using `concat` function and `inner` join

```python
# Join UK, France, and Netherlands full datasets by axis
full_df = pd.concat([france_df, netherlands_df, uk_df], axis="rows", join="inner")
full_df.head()
```

### Concatenate data by rows using `concat` function and `inner` join and make each country a key.

```python
# Join UK, France, and Netherlands full datasets by axis and add the countries to the index.
full_df = pd.concat([france_df, netherlands_df, uk_df], axis="rows", join="inner", keys=["France", "Netherlands", "UK"])


# Drop the Country column and display the updated DataFrame.
full_df = full_df.drop(columns="Country")
full_df.head()
```
