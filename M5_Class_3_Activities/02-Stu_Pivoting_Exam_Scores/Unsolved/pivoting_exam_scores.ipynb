```python
# Import Dependencies
import pandas as pd
```

```python
# Read file into DataFrame
exam_scores_df = pd.read_csv('Resources/students_exam_scores.csv')
exam_scores_df.head(10)


```

```python
# Using the `pivot_table()` function show average exam score for each subject.
# The subject should be the columns. Round to one decimal place.The subjects are 'Math', 'Speech', 'Composition', 'Biology'
average_scores = pd.pivot_table(exam_scores_df, columns=['subject'], values='exam_score').round(1)


# Rename the index: "Avg_Subject_Score" and round to one decimal place.
Avg_Subject_Score = average_scores.rename_axis(None, axis=1).rename_axis('Avg_Subject_Score', axis=0).round(1) 
Avg_Subject_Score.head(10)
```

```python
# Using the `pivot_table()` function show the minumum and maximum score for each subject. 
min_max_scores = pd.pivot_table(exam_scores_df, columns=['subject'], values='exam_score', aggfunc=['min', 'max']).round(1)
min_max_scores
```

```python
# Using the `pivot_table()` function show the average score for each student across all subjects. 
# Round to one decimal place.
average_student_score = pd.pivot_table(exam_scores_df, columns= "student_name", values='exam_score').round(1)


# Rename the index "Avg. Student Score"
Avg_Student_Score = average_student_score.rename_axis('Avg. Student Score', axis=0)
Avg_Student_Score
```

```python
# Using the `pivot_table()` function show the minimum and maximum exam score for each student across all subjects. 
student_min_max_scores = pd.pivot_table(exam_scores_df, columns= "student_name", values='exam_score', aggfunc=['min', 'max']).round(1)
student_min_max_scores


```

```python
# Using the `pivot_table` or the `pivot` function show each students exam score for each subject.
# The student names are the columns, the subject is the index, and the values are the exam scores.
student_subject_scores = pd.pivot_table(exam_scores_df, columns= "student_name", index='subject', values='exam_score').round(1)
student_subject_scores
```
