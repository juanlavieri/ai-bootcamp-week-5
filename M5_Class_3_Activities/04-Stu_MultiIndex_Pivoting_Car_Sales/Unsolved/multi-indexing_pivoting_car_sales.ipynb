```python
# Import dependencies
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
vehicles_df = pd.read_csv('Resources/vehicles.csv')
vehicles_df.head()
```

```python
# Show the total number of vehicles for each model and make.
grouped_vehicles = vehicles_df.groupby(['model', 'make']).size()

# Show the top 20 results.
grouped_vehicles.head(20)
```

```python
# Rename the "count" column "Total"
vehicles_df = vehicles_df.rename(columns={'count': 'Total'})


# Sort the renamed pivot table on the "Total" column to show the top 20 vehicles sold by model and make. 
vehicles_df = vehicles_df.sort_values('Total', ascending=False)
vehicles_df.head(20)
```

```python
# Show the total number of vehicles for each model, make, and year.
grouped_vehicles_by_year = vehicles_df.groupby(['model', 'make', 'year']).size()

# Show the top 20 results.
grouped_vehicles_by_year.head(20)
```

```python
# Rename the "count" column to "Total".
vehicles_df = vehicles_df.rename(columns={'count': 'Total'})


# Sort the pivot table on the "year" index and "Total" column and show the top 20 vehicles sold by the year.
vehicles_df = vehicles_df.sort_values(['year', 'Total'], ascending=[True, False])
vehicles_df.head(20)

```

```python
# Show the total number of vehicles for each category and year. 
grouped_vehicles_by_category = vehicles_df.groupby(['category', 'year']).size()

# Show the top 20 results.
grouped_vehicles_by_category.head(20)
```
