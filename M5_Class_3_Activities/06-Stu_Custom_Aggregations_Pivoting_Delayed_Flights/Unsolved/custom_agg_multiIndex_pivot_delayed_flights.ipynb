```python
# Import the necessary dependencies
import pandas as pd
```

```python
# Read in the data as a DataFrame
# Provide the dataset URL from an AWS S3 bucket. It's ~100 MB.
delayed_flights_url = "https://static.bc-edx.com/ai/ail-v-1-0/m5/lesson_2/delayed_flights.csv"
delayed_flights_df = pd.read_csv(delayed_flights_url)
delayed_flights_df.head()
```

```python
# Get the columns
delayed_flights_df.columns.value_counts()

```

```python
# Create a custom function that will calculate the average of DataFrame column
def custom_avg(column):
    """Calculate the average of a DataFrame column."""
    return column.mean()



# Create a pivot table to show the largest average arrival delay
pivot_avg_arrival_delay = delayed_flights_df.pivot_table(
    values='ArrDelay', 
    index=['UniqueCarrier', 'DayofMonth', 'DayOfWeek'], 
    aggfunc=custom_avg
)

# Sort the pivot table to show the top 25 results
pivot_top_25_arrival_delay = pivot_avg_arrival_delay.sort_values(by='ArrDelay', ascending=False).head(25)

pivot_top_25_arrival_delay
```

```python
# Use the custom_avg function to show the average delay of arrivals and departures
# for each carrier, day of the months and day of the week. Round to one decimal place. 

pivot_avg_arrival_departure_delay = delayed_flights_df.pivot_table(
    values=['ArrDelay', 'DepDelay'],
    index=['UniqueCarrier', 'DayofMonth', 'DayOfWeek'],
    aggfunc=custom_avg
)

# Rename the columns to reflect the aggregation. 


# Sort the pivot table to show the greatest average arrival and departures delays
# by carrier, day of the months and day of the week. Show the top 25 results
pivot_top_25_arrival_departure_delay = pivot_avg_arrival_departure_delay.sort_values(by=['ArrDelay', 'DepDelay'], ascending=False).head(25)

pivot_top_25_arrival_departure_delay
```

```python
# Show the bottom 25 results.
pivot_bottom_25_arrival_departure_delay = pivot_avg_arrival_departure_delay.sort_values(by=['ArrDelay', 'DepDelay'], ascending=True).head(25)

pivot_bottom_25_arrival_departure_delay
```

```python
# Create a custom function that calculates the total of a DataFrame column.
def custom_total(column):
    """Calculate the total of a DataFrame column."""
    return column.sum()
```

```python
# Create a pivot table to show the total and average number of flights that were diverted
# for each carrier, the origin, and destination
pivot_diverted_flights = delayed_flights_df.pivot_table(
    values='Diverted',
    index=['UniqueCarrier', 'Origin', 'Dest'],
    aggfunc={'Diverted': [custom_total, custom_avg]}
)

# Rename the columns to reflect the results
pivot_diverted_flights.columns = ['Total Diverted', 'Average Diverted']

# Sort the pivot table to show the greatest average and number of diverted flights
# for each carrier, the origin, and destination
pivot_top_25_diverted_flights = pivot_diverted_flights.sort_values(by=['Total Diverted', 'Average Diverted'], ascending=False).head(25)

pivot_top_25_diverted_flights
```
